## Correlation graphs


```python
 from math import isnan
def correlation(a,b):
    ocm=a.reshape(-1)
    modis=b.reshape(-1)

   
    x1=[]
    y1=[]
    for oi,mi in zip(ocm,modis):
        if(oi<1 and mi<1 and isnan(oi)==0 and isnan(mi)==0  ):
            x1.append(oi)
            y1.append(mi)
    x=np.asarray(x1)
    y=np.asarray(y1)
    coef2=np.polyfit(x,y,1)
    return x,y,coef2
```

```python
aod_18_1,aod_19_1,cor_18_19_1= correlation(aod_2018_1,aod_2019_1)
aod_18_2,aod_19_2,cor_18_19_2= correlation(aod_2018_2,aod_2019_2)
aod_19_1_b,aod_20_1,cor_19_20_1= correlation(aod_2019_1,aod_2020_1)
aod_19_2_b,aod_20_2,cor_19_20_2= correlation(aod_2019_2,aod_2020_2)
```

## Time Sereis
-----------
#### To demonstrate that the AOD values have decreased for the lockdown months of, A time series forecast model is prepared, that proves that the forecasted values without the assumption of lockdown is higher than the actual values. The study area chosen for this exercise is Delhi.

```python
from statsmodels.tsa.arima_model import ARIMA
import pmdarima as pm
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.graphics.tsaplots import plot_pacf
```

```python
df18=pd.read_csv(r"D:\BSA_assignment\MODIS_AOD\delhi\aod_2018.csv")
df19=pd.read_csv(r"D:\BSA_assignment\MODIS_AOD\delhi\aod_2019.csv")
df20=pd.read_csv(r"D:\BSA_assignment\MODIS_AOD\delhi\aod_2020.csv")
dates18=df18["dates"]
dates19=df19["dates"]
data18=df18["AOD"]
data19=df19["AOD"]
dates20=df20["dates"]
data20=df20["AOD"]
```

```python
from math import isnan
dates_tr=[]
data_tr=[]

for i,j in zip(dates18,data18):
    if(isnan(j)==0):
        dates_tr.append(i)
        data_tr.append(j)
for i,j in zip(dates19,data19):
    if(isnan(j)==0):
        dates_tr.append(i)
        data_tr.append(j) 
for i in range(10):
    
        dates_tr.append(dates20[i])
        data_tr.append(data20[i]) 
```

```python
smodel = pm.auto_arima(data_tr, start_p=1, start_q=1,
                         test='adf',
                         max_p=3, max_q=3, m=4,
                         start_P=0, seasonal=True,
                         d=None, D=1, trace=True,
                         error_action='ignore',  
                         suppress_warnings=True, 
                         stepwise=True)

smodel.summary()
```

```python
data_truth=data20[11:]
dates_truth=dates20[11:]
```

```python
# Forecast
traindata=pd.DataFrame()
traindata["AOD"]=data_tr
traindata["dates"]=dates_tr
n_periods = 16
fitted, confint = smodel.predict(n_periods=n_periods, return_conf_int=True)
```

```python
plt.figure(figsize=(45,15))
plt.plot(dates_tr,data_tr,label="Training data")
plt.plot(dates_truth,data_truth,label="Actual data")
plt.plot(dates_truth,fitted,label="forecasted")
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

The orange line depicts the actual value, the blue line depicts the training data and the green line depicts the forecasted values. It is clearly seen in the figure that actual values for most of the weeks were lower than the forecasted values. Hence, it shows that stopping of various anthropogenic activities affects the pollution content in the region.

```python
def forecast_accuracy(forecast, actual):
    mape = np.mean(np.abs(forecast - actual)/np.abs(actual))  # MAPE
    me = np.mean(forecast - actual)             # ME
    mae = np.mean(np.abs(forecast - actual))    # MAE
    mpe = np.mean((forecast - actual)/actual)   # MPE
    rmse = np.mean((forecast - actual)**2)**.5  # RMSE
    corr = np.corrcoef(forecast, actual)[0,1]   # corr
    mins = np.amin(np.hstack([forecast[:,None], 
                              actual[:,None]]), axis=1)
    maxs = np.amax(np.hstack([forecast[:,None], 
                              actual[:,None]]), axis=1)
    minmax = 1 - np.mean(mins/maxs)             # minmax
    #acf1 = acf(fc-test)[1]                      # ACF1
    return({'mape':mape, 'me':me, 'mae': mae, 
            'mpe': mpe, 'rmse':rmse,
            'corr':corr, 'minmax':minmax})
```

```python
forecast_accuracy(fitted,data_truth)
```
